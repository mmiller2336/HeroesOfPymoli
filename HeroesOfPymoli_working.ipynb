## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels=group_labels)

purchase_total = purchase_data.groupby(["Age Ranges"]).sum()["Price"].rename("Total Purchase Value")
avg_purchase_price = purchase_data.groupby(["Age Ranges"]).mean()["Price"].rename("Average Purchase Price")
purchase_count = purchase_data.groupby(["Age Ranges"]).count()["Price"].rename("Purchase Count")

purchase_data.head()

purchase_analysis_df = pd.DataFrame({"Total Purchase Value": purchase_total, "Average Purchase Price": avg_purchase_price, "Purchase Count": purchase_count })

purchase_analysis_df

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_count = purchase_data.groupby(["SN"]).sum()["Price"]
avg_purchase_price = purchase_data.groupby(["SN"]).mean()["Price"]
total_sn = purchase_data.groupby(["SN"]).count()["Price"]
total_purchase_value = purchase_count / total_sn



summary_table = pd.DataFrame({"Purchase Count": purchase_count, "Avg Purchase Price": avg_purchase_price, "Total Purchase Value": total_purchase_value})
summary_table

summary_table.sort_values(by="Total Purchase Value", ascending=False)
```
